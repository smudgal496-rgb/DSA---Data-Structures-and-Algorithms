```python
def infix_to_postfix(exp):
    stack=[]
    output=[]
    priority={"**":3, "$":3, "^":3, "/":2, "*":2, "+":1, "-":1}

    for char in exp:
        if char.isalnum():
            output.append(char)

        elif char == "(":
            stack.append(char)
        elif char == ")":
            while stack and stack[-1] != "(":
                output.append(stack.pop())
            stack.pop()
        else:
            while stack and stack[-1] != "(" and priority.get(char,0) <= priority.get(stack[-1],0):
                output.append(stack.pop())
            stack.append(char)

    while stack:
        output.append(stack.pop())

    return "".join(output)

new=(infix_to_postfix("9+(1*2-(3/5^8)*2)*1"))
print (new)
```

```python
def post_exp_eval(exp):
    stack = []
    tokens = exp.split()   # split by space

    for token in tokens:
        if token.isdigit():
            stack.append(int(token))
        else:
            top_el = stack.pop()
            nex_el = stack.pop()

            if token == '+':
                stack.append(nex_el + top_el)
            elif token == '-':
                stack.append(nex_el - top_el)
            elif token == '*':
                stack.append(nex_el * top_el)
            elif token == '/':
                stack.append(nex_el / top_el)
            elif token == '^':
                stack.append(nex_el / top_el)
            else:
                raise ValueError(f"Unsupported operator: {token}")

    return stack.pop()

result = post_exp_eval("9 1 2 * 3 5 8 ^ / 2 * - 1 * +")
print(result)
```
